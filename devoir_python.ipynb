```python
# Cellule 1 — Partie (a) : Gradient
import sympy as sp
x, y = sp.symbols('x y')
f = x**3 + y**3 - 3*x*y
df_dx = sp.diff(f, x)
df_dy = sp.diff(f, y)
print(df_dx, df_dy)
```

```python
# Partie (b) : Jacobienne
gradient = sp.Matrix([df_dx, df_dy])
jacobienne = gradient.jacobian([x, y])
print("Matrice Jacobienne :")
sp.pprint(jacobienne)
```

```python
# Partie (c) : Hessienne
hessienne = sp.hessian(f, [x, y])
print("Matrice Hessienne :")
sp.pprint(hessienne)
```

```python
# Partie (d) : Points critiques
points_critiques = sp.solve([df_dx, df_dy], [x, y])
print("Points critiques :", points_critiques)
```

```python
# Partie (e) : Classification (version corrigée)
for pt in points_critiques:
    px, py = pt
    
    # Ignorer les points complexes
    if not (px.is_real and py.is_real):
        print(f"\nPoint ({px}, {py}) : complexe → ignoré")
        continue
    
    H = hessienne.subs([(x, px), (y, py)])
    det_H = H.det()
    H11 = H[0, 0]

    print(f"\nPoint ({px}, {py}) :")
    sp.pprint(H)
    print(f"det(H) = {det_H}")
    print(f"H[0,0] = {H11}")

    if det_H < 0:
        print("→ Point selle ⚠️")
    elif det_H > 0 and H11 > 0:
        print("→ Minimum local ✅")
    elif det_H > 0 and H11 < 0:
        print("→ Maximum local ✅")
```
